```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns 
from mpl_toolkits.basemap import Basemap
%matplotlib inline

plt.figure(1, figsize=(12,6))

# Mercator of World
m1 = Basemap(projection='merc',
             llcrnrlat=-60,
             urcrnrlat=65,
             llcrnrlon=-180,
             urcrnrlon=180,
             lat_ts=0,
             resolution='c')

m1.fillcontinents(color='#191919',lake_color='#000000') # dark grey land, black lakes
m1.drawmapboundary(fill_color='#FFFFFF')                # black background
m1.drawcountries(linewidth=0.1, color="w")              # thin white line for country borders

plt.title("Peta Dunia")
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(1, figsize=(19,13))

# Sample it down to only the Java Island
lon_min, lon_max = 105.21, 114.60527777778
lat_min, lat_max = -8.78, -5.8758333333333

# Mercator of Java Island
m1 = Basemap(projection='merc',
             llcrnrlat=lat_min,
             urcrnrlat=lat_max,
             llcrnrlon=lon_min,
             urcrnrlon=lon_max)

m1.drawmapboundary(fill_color='grey', linewidth=0)
m1.fillcontinents(color='white',lake_color='black')
m1.drawcoastlines(linewidth=0.1, color="white")

lons = [112.7373, 112.4136, 111.3428, 111.8967, 111.5257, 111.0390, 111.1053, 110.8120, 110.0062, 109.6749, 108.3216, 108.5521, 107.3002, 107.6008, 106.8317]
lats = [-7.2709, -7.1182, -6.7042, -7.6068, -7.6334, -6.7523, -8.2013, -7.5622, -7.7126, -6.8875, -6.3304, -6.7191, -6.2938, -6.9201, -6.1570]
labels = ['Surabaya','Lamongan','Rembang','Nganjuk','Madiun','Pati','Pacitan','Surakarta','Purworejo','Pekalongan','Indramayu','Cirebon','Karawang','Bandung','Jakarta']

lons1 = [111.7127, 110.3785, 109.6662, 107.8996, 106.9213]
lats1 = [-8.0524, -7.7898, -7.6753, -7.2284, -6.9214]
labels1 = ['Trenggalek','Yogjakarta','Kebumen','Garut','Sukabumi']

x_offsets = [10000, -30000, -25000]
y_offsets = [5000, -25000, -25000]
x1_offsets = [10000, -30000, -25000]
y1_offsets = [5000, -25000, -25000]

x,y = m1(lons, lats)
x1,y1 = m1(lons1, lats1)
m1.plot(x, y, 'bo', c="blue", markersize=10)
m1.plot(x1, y1, 'bo', c="red", markersize=10)

plt.title("Pemetaan Titik Daerah Rawan Bencana di Pulau Jawa")

for label, x, y in zip(labels, x, y):
    plt.text(x+10000, y+5000, label)
for label1, x1, y1 in zip(labels1, x1, y1):
    plt.text(x1+10000, y1+5000, label1)
    
plt.show()
```
